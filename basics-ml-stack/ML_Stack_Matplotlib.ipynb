# ML Stack Matplotlib

## Points

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)

y = np.sin(x)
```

```
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("fn")
plt.title("My Chart")
plt.show()
```

```
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

## Scatterplot

```
import pandas as pd

# https://github.com/lazyprogrammer/machine_learning_examples/raw/master/linear_regression_class/data_1d.csv
A = pd.read_csv("data_1d.csv", header=None).to_numpy()
A[:5]
```

```
x = A[:,0] # all rows of 0 column
y = A[:,1] # all rows of 1 column
plt.scatter(x, y)

x_line = np.linspace(0, 100, 100)
y_line = 2 * x_line + 1
plt.plot(x_line, y_line)

plt.show()
```

## Histogram

```
plt.hist(x)
plt.show()
```

```
R = np.random.random(10_000)
plt.hist(R)
plt.show()
```

```
y_actual = 2 * x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```
