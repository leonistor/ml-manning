# MatPlotLib Basics

```
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
plt.show()
```

## Multiple plots

```
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

## Adjust the axes, add grid, line type and color, label axes

```
axes = plt.axes()

axes.set_xlim([-5, 5])
axes.set_ylim([0, 1.0])

axes.set_xticks(list(range(-5,6)))
axes.set_yticks(np.linspace(0, 1, 11))

axes.grid()

plt.xlabel('Frobbles')
plt.ylabel('Value')

plt.plot(x, norm.pdf(x), 'b-') # blue solid
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:') # red, dashed
plt.show()
```

## Scatter plot

```
X = np.random.randn(500)
Y = np.random.randn(500)
plt.scatter(X, Y)
plt.show()
```

## Box and whisker plot

Visualize spread & skew of data:

- red is median
- box from 1st to 3rd quartiles
- 50% of data in the box
- dotted line whiskers -> range of data
- except outliers (outside 1.5 x inter-quartile-range, i.e. the size of the box)

```
unifrom_skewed = np.random.rand(100) * 100 -40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100

data = np.concatenate((unifrom_skewed, high_outliers, low_outliers))

plt.boxplot( data)
plt.show( )
```
